```python
import requests
import shutil
```

```python
locale ="en"
searchText = "dog"
url = f"https://api.arasaac.org/api/pictograms/{locale}/search/{searchText}"
```

```python
result = requests.get(url.format(locale,searchText))
```

```python
json = result.json()
id = json[0]["_id"]
id
```

```python
download_url = f'https://api.arasaac.org/api/pictograms/{id}?download=true'
response = requests.get(download_url, stream=True)

with open(f'img/img_{id}.png', 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)
del response

```
